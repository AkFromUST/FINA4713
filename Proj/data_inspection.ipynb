```python
import pandas as pd
import numpy as np

# Load the data
df = pd.read_parquet('jkp_data.parquet')

print("="*80)
print("JKP DATA INSPECTION")
print("="*80)

# Basic dimensions
print(f"\n1. BASIC SHAPE:")
print(f"   Rows: {len(df):,}")
print(f"   Columns: {len(df.columns)}")

# Date range
print(f"\n2. TIME RANGE:")
print(f"   Start: {df['eom'].min()}")
print(f"   End: {df['eom'].max()}")
print(f"   Total months: {df['eom'].nunique():,}")
```

```python
print(f"\n3. STOCK COUNT:")
print(f"   Unique stocks: {df['id'].nunique():,}")
print(f"   Avg stocks per month: {df.groupby('eom').size().mean():.0f}")

# Target variable
print(f"\n4. TARGET VARIABLE (ret_exc_lead1m):")
print(f"   Mean: {df['ret_exc_lead1m'].mean():.6f}")
print(f"   Std: {df['ret_exc_lead1m'].std():.6f}")
print(f"   Min: {df['ret_exc_lead1m'].min():.6f}")
print(f"   Max: {df['ret_exc_lead1m'].max():.6f}")
print(f"   Sample (first 10): {df['ret_exc_lead1m'].head(10).values}")

# Feature columns
exclude = ['id', 'eom', 'excntry', 'ret_exc_lead1m', 'me', 'mcap_usd']
feature_cols = [c for c in df.columns if c not in exclude]
print(f"\n5. FEATURES:")
print(f"   Number of features: {len(feature_cols)}")
print(f"   First 10 features: {feature_cols[:10]}")

# Missing values
print(f"\n6. MISSING VALUES (%):")
missing_pct = df[feature_cols].isna().mean() * 100
print(f"   Mean missing: {missing_pct.mean():.2f}%")
print(f"   Median missing: {missing_pct.median():.2f}%")
print(f"   Max missing: {missing_pct.max():.2f}%")
print(f"   Features with 0% missing: {(missing_pct == 0).sum()}")
print(f"   Features with >50% missing: {(missing_pct > 50).sum()}")
```
